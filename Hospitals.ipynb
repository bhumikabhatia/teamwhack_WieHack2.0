```python
def function():
    li=['Disease']
    n=int(input())
    for i in range(n):
        symp=input()
        li.append(symp)

    df2=df[li]   
    #print(df2)
    maxi=0
    for i in range(df.shape[0]):
        count=0
        for j in range(1,df2.shape[1]):
            if(df2.iloc[i,j]==1):
                count+=1
        #print(df2.iloc[i,0],count)
        if(count>maxi):
            maxi=count
            recom=df2.iloc[i,0]
    return (recom)
```
